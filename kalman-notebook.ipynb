# Kalman Filter

A jupyter notebook to use for testing and tweaking my implementation of the kalman filter.

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import datetime
from pandas_datareader import data, wb
import numpy as np
import pandas as pd
import kalman
```

```python
import quandl

quandl.ApiConfig.api_key = 'api key here'

dow_code = 'BCB/UDJIAD1'
sp_code = 'MULTPL/SP500_REAL_PRICE_MONTH'
yale_code = 'YALE/SPCOMP'
monthly_invest = 500
start = '2000-02-12'
end = '2018-02-12'

sp_data = quandl.get(yale_code, returns="numpy", start_date=start, end_date=end)
# format: year, S&P Comp price, dividend, earnings, CPI, long interest rate, 
#    real price, real dividend, real earnings, cyclically adjusted PE 
close_prices = np.array([d[1] for d in sp_data])
dates = [d[0] for d in sp_data]

# using S&P Composite prices right now
kalman_filtered = kalman.kalman_filter(close_prices)
closes = plt.plot(dates, close_prices, label='close prices')
kalmans = plt.plot(dates,kalman_filtered, label='predicted prices')
plt.show()
```
